```python
import pandas as pd
import openpyxl as op
import pyarrow as pya
```

Tópico 1


```python
df = pd.read_excel("dados_a_tratar.xlsx")
print(df)
```

```python
print(df.isnull().sum())
```

2. Colunas categoricas

```python
df["Nome"] = df["Nome"].fillna("Desconhecida")
df["Departamento"] = df["Departamento"].fillna("Desconhecido")
print(df)
```

2. Colunas numéricas


```python
df["Idade"] = df["Idade"].replace("trinta", 30)
df["Idade"] = pd.to_numeric(df["Idade"])
df["Idade"] = df["Idade"].fillna(df["Idade"].mean())
df["Salário"] = df["Salário"].fillna(df["Salário"].mean()).round(2)
print(df)
```

2. Data_Admissao


```python
df["Data_Admissao"] = pd.to_datetime(df["Data_Admissao"], errors='coerce')
df["Data_Admissao"] = df["Data_Admissao"].fillna(pd.Timestamp("2000-01-01"))
print(df)
```

3. Remover duplicata

```python
print("Antes da filtragem de duplicatas: ", df.duplicated().sum())

df_sem_duplicata = df.drop_duplicates()
print(df_sem_duplicata)

print("Depois da filtragem de duplicatas: ", df_sem_duplicata.duplicated().sum())
```

4. Verificação do e-mail

```python
df_sem_duplicata["Email"] = df_sem_duplicata["Email"].apply(lambda x: x if "@" in str(x) and ".com" in str(x) else "email_inválido@email.com")

print(df_sem_duplicata["Email"])
```

6. Tipo de dados correto

```python
df_sem_duplicata["Idade"] = df_sem_duplicata["Idade"].astype(int)
df_sem_duplicata["Salário"] = df_sem_duplicata["Salário"].astype(float)
df_sem_duplicata["Data_Admissao"] = pd.to_datetime(df_sem_duplicata["Data_Admissao"])

df.info()
```

DATA FRAME

Quantidade de funcionários por departamento

```python

df_agrupado = pd.DataFrame(df_sem_duplicata.groupby("Departamento")["ID"].count().reset_index())


df_agrupado = df_agrupado.rename(columns={"ID": "Total_Funcionarios"})

print(df_agrupado)
```

Media dos salários de cada departamento

```python
media_dep = df_sem_duplicata.groupby("Departamento")["Salário"].mean().reset_index().round(2)

print(media_dep)
```

Funcionarios admititos depois de 2020

```python
admitidos = df_sem_duplicata[df["Data_Admissao"] > "2020-01-01"]

print(admitidos)
```

Tres funcionários mais bem pagos

```python
pagamento = df_sem_duplicata.sort_values("Salário", ascending=False)

pagamento.head(3)
```

Salario e idade média

```python
media_salario_idade = df.groupby("Departamento").agg({
    "Idade": "mean",
    "Salário": "mean"
}).reset_index().round(2)

print(media_salario_idade)
```

    Quantos funcionários foram admitidos

```python
df["Ano"] = df["Data_Admissao"].dt.year

admitidos_por_ano = df["Ano"].value_counts().sort_index().reset_index()
admitidos_por_ano.columns = ["Ano", "Quantidade"]

print(admitidos_por_ano)
```
